## 3. ***Data Wrangling***

### Data Wrangling Code

#### Removing Duplicate Value

````python
# Write your code to make your dataset analysis ready.
# Remove duplicates
df = df.drop_duplicates()
df.shape                  # after removing duplicate rows
````

We check the duplicate values there are 31994 duplicte rows. After removing duplicate rows there are 87396 rows and 32 columns.


#### Working on null values


We checked null values in our dataset from that we found that there are null values in columns - company, agent, country and children.

Let's take the comapny column

A) For company column it contains 94% null data.

````python
# For company column it contains 94% null data, so it is better to drop the company column.
df.drop(['company'],axis = 1,inplace = True)
````

````python
# after removing company column there are 31 columns remaining.
df.shape
````

Let's take agent column.

B) For agent column it contains 13.95% null data, so assuming those rooms were booked without any agents.

````python
df['agent'].mode()
# In missing value / null value we see that for agent column minimum number of null values are present. So we replace null values with mode value.
df['agent'].fillna(0, inplace = True)
````

````python
df['agent'].isnull().sum()
````

Let's  take country column.

C) For country it contains 0.52% null data , so assuming tourist belong to country other than available list.

````python
#from above we see that percentage of null values in country column is very less so we replaced the null values by Other
df['country'].fillna('other', inplace = True)
````

````python
df['country'].isnull().sum()
````

Let's take children column.

D) For children there are only 4 null value count, so assuming 0 children in that family.

````python
#from above we see that children column has very less null vaues therefore fill null values with 0
df['children'].fillna(value=0,inplace=True)
````

````python
df.isnull().sum()
````

#### Change the dtype of some variables.

````python
df.info()
````

````python
#from above info it is seen that the children column and agent column has float value which is converted into int64 for our requirnment
df[['children','agent']] = df[['children','agent']].astype('int64')
````

#### Creating the new columns for our requirement

````python
# create new column for total number of guest.
df['total_guest'] = df['adults'] + df['children'] + df['babies']
````

````python
# create new column for total stays in nights
df['total_stays_in_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
````

````python
# create new column for revenue of hotel
df['revenue'] = df['total_stays_in_nights'] * df['adr']
````

````python
df.info()
````

#### Converting Columns As Our Requirement

````python
#for our understanding replace the is_calceled column value of(0,1) with 'not canceled' and 'canceled'
df['is_canceled'] = df['is_canceled'].replace([0,1],['not canceled','is canceled'])
````

````python
#for our understamding replace the is_repeated_guest column of value (0,1) with the 'repeat' and 'not repeat'
df['is_repeated_guest'] = df['is_repeated_guest'].replace([0,1], ['not repeated','repeated'])
````

````python
# Chart 4 :

# groupby arrival_date_month and taking the hotel count
bookings_by_months_data = df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':"Counts"})
# Creating list of months in order
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
# creating df which will map the order of above months list without changing its values.
bookings_by_months_data['arrival_date_month'] = pd.Categorical(bookings_by_months_data['arrival_date_month'],categories=months,ordered=True)
# sorting by arrival_date_month
bookings_by_months_data = bookings_by_months_data.sort_values('arrival_date_month')

bookings_by_months_data
````

````python
# Visulisation code

# Setting plot size
plt.figure(figsize=(15,5))

# Plotting lineplot on x-months & y-booking counts
sns.lineplot(x=bookings_by_months_data['arrival_date_month'],y=bookings_by_months_data['Counts'])

# Setting title for the plot
plt.title('Number of bookings across each month',fontsize=25)
# Setting x label
plt.xlabel('Month',fontsize=20)
# Setting y label
plt.ylabel('Number of bookings',fontsize=20)
````
